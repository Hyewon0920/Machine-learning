### WR rate에 가중치를 주는 법


```python
#전체 영화의 평균평점
C = df2['vote_average'].mean()
C
```

```python
# 하위 10%에 대한 데이터 뽑음
m = df2['vote_count'].quantile(0.9)
m
```

```python
# df2를 복사해서 사용할 거임, 단 하위10%보다 크거나 같은 애들만
q_movies = df2.copy().loc[df2['vote_count'] >= m]
q_movies.shape
```

```python
q_movies['vote_count'].sort_values()
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v / (v + m) * R) + (m / (m + v) * C)
```

```python
# score라는 새 칼럼을 만드는 것
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
q_movies.head(3)
```

```python
q_movies = q_movies.sort_values('score', ascending=False)
q_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
# 영화의 제목을 입력받으면 코사인 유사도를 통해서 가장 유사도가 높은 상위 10개의 영화 목록 반환
def get_recommendations(title, cosine_sim=cosine_sim):
    # 영화 제목을 통해서 전체 데이터 기준 그 영화의 index 값을 얻기
    idx = indices[title]
    
    # 코사인 유사도 매트릭스 (cosine_sim) 에서 idx 에 해당하는 데이터를 (idx, 유사도) 형태로 얻기
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # 코사인 유사도 기준으로 내림차순 정렬
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # 자기 자신을 제외한 10개의 추천 영화를 슬라이싱
    sim_scores = sim_scores[1:11]
    
    # 추천 영화 목록 10개의 인덱스 정보 추출
    movie_indices = [i[0] for i in sim_scores]
    
    # 인덱스 정보를 통해 영화 제목 추출
    return df2['title'].iloc[movie_indices]
```

```python
# 처음 3개의 데이터 중에서 name 에 해당하는 value 만 추출
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        if len(names) > 3:
            names = names[:3]
        return names
    return []
```

```python
features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(' ', '')) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(' ', ''))
        else:
            return ''
```

```python
features = ['cast', 'keywords', 'director', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df2['soup'] = df2.apply(create_soup, axis=1)
df2['soup']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df2['soup'])
count_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
cosine_sim2
```

# 협업 필터링

```python
import surprise
surprise.__version__
```

```python
import pandas as pd
from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate
```

```python
ratings = pd.read_csv('ratings_small.csv')
ratings.head()
```

```python
ratings['rating'].min()
```

```python
ratings['rating'].max()
```

```python
reader = Reader(rating_scale=(0.5, 5))
```

```python
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader=reader)
data
```

```python
svd = SVD(random_state=0)
```

```python
cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```
