# 지도, 비지도

지도 : 일정 라벨링을 정해놓고 머신러닝에 들어감 -> 이 라벨링을 통해 특정 라벨링의 특징을 이해하고 다른것도 구분할 수 있게됨

비지도 : 라벨링없음, 정답없음, 스스로 어떤 기준과 패턴을 찾고 무리를 지음

강화학습 : 게임에서 자주씀, 어떤 환경을 주어주고 적절한 보상, 패널티를 주면서 학습하게 함 -> 최종적으로 보상을 가장 많이 받는 방법을 찾아냄


# 지도학습

- 회귀
 - 변수들 간의 상관관계를 찾는 것, 연속적인 데이터로부터 결과 예측 
 - 예측변수가 숫자일 때 사용하기 좋음
 - 이와 같이 연속적인 숫자 데이터를 통해 예측을 하는 것을 '회귀 모델' 이라고 함

- 분류
 - 어떤 영화가 좋은지 나쁜지는 감독, 장르 정도가 영향을 줄 수 있음
 - 
 - 새로운 영화 G의 코믹, 폭력성이 높다면 좋아한다 쪽으로 넘어갈 수 있을 것이라고 예측할 수 있음
 - 정해진 범주에 넣어놓는 것, 숫자가 아닐 때 사용하기 유용함
